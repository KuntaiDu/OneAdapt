```python
from PIL import Image
import torchvision.transforms as T
from dnn.dnn_factory import DNN_Factory
```

```python
orig = Image.open('example.jpg').resize((640, 480))
image = T.ToTensor()(orig)
print(image.shape)

model = DNN_Factory().get_model('COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml')
```

Image.open('example_dog.jpg').resize((512, 512)

```python

gamma = 1.0

ret = T.ToPILImage()(T.functional.adjust_gamma(image,gamma=gamma, gain=1.2))


result = model.inference( T.functional.adjust_gamma(image,gamma=gamma, gain=1.2).unsqueeze(0).cuda(), detach=True)

model.visualize(ret, result)
```
